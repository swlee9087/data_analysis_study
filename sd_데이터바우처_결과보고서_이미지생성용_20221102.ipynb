```python
# 작성자 나이별 집계
final.groupby("글작성자나이")["글작성자나이"].count()
```

```python
# 카테고리별 작성자 나이 집계
cat_age = final.groupby(['카테고리', '글작성자나이']).size().unstack()
cat_age['total'] = cat_age.sum(axis=1)
cat_age.sort_values('total', inplace=True)
cat_age
```
